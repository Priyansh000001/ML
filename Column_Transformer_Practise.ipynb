```python
cat_columns = ['Gender', 'Marital Status', 'Education Level', 'Employment', 'Type of Loan']
num_columns = df.select_dtypes(exclude='object').columns.tolist()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder, StandardScaler

transformer = ColumnTransformer(
    transformers=[
        ('gender_en', OrdinalEncoder(categories=[['Male', 'Female']]), ['Gender']),
        ('marital_en', OrdinalEncoder(categories=[['Married', 'Divorced', 'Single']]), ['Marital Status']),
        ('education_en', OrdinalEncoder(categories=[['High School', 'Bachelor', 'Master', 'PhD']]), ['Education Level']),
        ('emp_en', OrdinalEncoder(categories=[['Unemployed', 'Employed', 'Self-Employed']]), ['Employment Status']),
        ('loan_en', OrdinalEncoder(categories=[['Auto Loan', 'Home Loan', 'Personal Loan']]), ['Type of Loan']),
        ('scaler', StandardScaler(), num_columns)
    ],
    remainder='drop'
)

df_transformed = transformer.fit_transform(df)
```

```python
df_transformed = transformer.fit_transform(df)

feature_names = transformer.get_feature_names_out()
df_transformed = pd.DataFrame(df_transformed, columns=feature_names)
```
